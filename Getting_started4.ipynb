```python
# Linear Algebra
# We do matrix multiplication here

c = np.full((2,3),2)
print(c)

d = np.ones((3,2))
print(d)

np.matmul(c,d)
```

```python
# Determinant of matrix
e = np.identity(3)

np.linalg.det(e)
```

```python
# Statistics

stats = np.array([[1,2,3],[4,5,6]])
stats
```

```python
np.min(stats,axis=1)
# It gives min value at first row, second row and so on
# Here min value at first row is 1 and that of at second row is 4
```

```python
np.min(stats)
# It gives minumum value from whole matrix.
```

```python
np.max(stats)
# It gives maximum value from whole matrix.
```

```python
np.max(stats,axis = 1)
# It gives maximum value at first row, second row and so on.
# Here max value at first row is 3 and that of at second row is 6
```

```python
# Reorganizing arrays

before = np.array([[1,2,3,4],[5,6,7,8]])
print(before.shape)

after = before.reshape((2,4))
print(after)
```

```python
# Vertically stacking vectors

v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])
np.vstack([v1,v2,v1,v2])
```

```python
# Horizontally stacking vectors

h1 = np.ones((2,4))
h2 = np.zeros((2,2))
np.hstack([h1,h2])
```

```python
import numpy as np
```

```python
## All 0s martix
np.zeros((5,3,8))
```

```python
# All 1s matrix
np.ones((4,2,2), dtype='int32')
```

```python
# Any other number
np.full((2,3),99, dtype='float32')
```

```python
# Any other number (full_like)

a = np.array([1,2,3]),([4,5,6])
np.full_like(a,4)
```

```python
# Random decimal numbers between 0s and 1s
np.random.rand(4,2,3)
```

```python
# Random Integer Values

np.random.randint(-4,9,size=(3,3))
```

```python
# Identity Matirx

np.identity(5)
```

```python
# Repeating the matrix

arr= np.array([[1,2,3]])
r1 = np.repeat(arr,3,axis=0)
print(r1)
```

```python
output = np.ones((5,5))
print(output)

z = np.zeros((3,3))
z[1,1] = 9
print(z)

output[1:-1,1:-1] = z
print(output)
```

```python
# Things we should notice while copying one array to another array

c = np.array([1,2,3])
d = c.copy()
d[0] = 100

print(d)
print(c)
```
